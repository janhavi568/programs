```python
import pandas as pd

data = {
    "Client_ID":[1001,1002,1003,1001,1004],
    "Freight_Cost":[4500,"missing",3100,2800,1500],
    "Jan_Tonnage":[12000,8500,15000,9000,11000],
    "Feb_Tonnage":[13500,9200,16200,9800,11500],
    "Sensor_Temp":[21.5,23.1,22.0,21.8,24.5]
}

df = pd.DataFrame(data)

# Task 2.1
df["Freight_Cost"] = df["Freight_Cost"].replace("missing", 0)

# Task 2.2
df["Freight_Cost"] = df["Freight_Cost"].fillna(0)

# Task 2.3
df["Freight_Cost"] = df["Freight_Cost"].astype(int)

print(df)
print("\nData Types:")
print(df.dtypes)
```

```python
mean_value = df["Freight_Cost"].mean()
median_value = df["Freight_Cost"].median()

print("Mean =", mean_value)
print("Median =", median_value)
```

```python
import pandas as pd
```

```python
df = pd.read_csv("data2.csv")
```

```python
print(df.info())

print("\nData Types:")
print(df.dtypes)

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
mean = df["Spend_Amount"].mean()
median = df["Spend_Amount"].median()
skewness = df["Spend_Amount"].skew()

print("Mean:", mean)
print("Median:", median)
print("Skewness:", skewness)
```

```python
pivot = df.pivot_table(
    values="Satisfaction_Score",
    index="Region",
    columns="Website_Version",
    aggfunc="mean"
)

print(pivot)
```

```python
from scipy import stats

old = df[df["Website_Version"]=="Old_Layout"]["Spend_Amount"]
new = df[df["Website_Version"]=="New_Layout"]["Spend_Amount"]

t_stat, p_value = stats.ttest_ind(old, new, equal_var=False)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)
```

```python
if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("bazaar_assignment.db")

df = pd.read_csv("data2.csv")

df.to_sql(
    "bazaar_ledger",
    conn,
    if_exists="replace",
    index=False
)
```

```python
query = """
SELECT Customer_ID,
       Name,
       Spend_Amount
FROM bazaar_ledger
WHERE Spend_Amount >= 4000;
"""

result = pd.read_sql_query(query, conn)
print(result)
```

```python
query = """
SELECT Region,
       SUM(Spend_Amount) AS Total_Revenue
FROM bazaar_ledger
GROUP BY Region;
"""

result = pd.read_sql_query(query, conn)
print(result)
```
